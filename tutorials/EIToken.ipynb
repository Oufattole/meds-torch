```python
from meds_torch import EIToken

data_path = "../tests/test_data/final_cohort/train/0.parquet"
df = EIToken.load_data(data_path)
print("\nInitial DataFrame:")
print(df)
```

```python
# Assign quantiles to every code with a numerical_value

# Quantile boundaries for codes with numerical values
quantile_df = EIToken.calculate_quantiles(df)
print("\nQuantile boundaries for every code in DataFrame:")
print(quantile_df)

# Assign quantiles
df = EIToken.assign_quantiles(df)
print("\nDataFrame after assigning quantiles:")
print(df)
```

```python
# Create event tokenizations with both joint and split (code, quantile) pairs
df = EIToken.create_event_tokens(df)
print("\nDataFrame after creating event tokens:")
print(df)
```

```python
# Calculate time intervals
df = EIToken.calculate_time_intervals(df)
print("\nDataFrame after calculating time intervals:")
print(df)
```

```python
# Generate patient timelines
timeline_df = EIToken.generate_patient_timeline(df)
print("\nDataFrame with generated patient timeline:")
print(timeline_df)
```

```python
import random
from datetime import datetime, timedelta

start_timestamp = datetime(2022, 1, 1, 12, 0)

# Define a list of potential intervals in terms of minutes, hours, days, and months
intervals = [
    timedelta(minutes=random.randint(1, 60)),
    timedelta(hours=random.randint(1, 24)),
    timedelta(days=random.randint(1, 7)),
    timedelta(days=random.randint(8, 30)),
    timedelta(days=30 * random.randint(1, 3)),
    timedelta(days=30 * random.randint(3, 6)),
    timedelta(days=30 * random.randint(6, 12)),
]

# Generate timestamps
timestamps = [start_timestamp]

timestamps = [start_timestamp]
for _ in range(21):  # Generate additional 21 timestamps
    next_interval = random.choice(intervals)
    timestamps.append(timestamps[-1] + next_interval)


print(timestamps)
```
